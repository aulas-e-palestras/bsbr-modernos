## Importações

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from itertools import combinations_with_replacement
```

## Dados

```python
df = (
    pd.read_csv('../data/data_svm.csv',usecols=['x','z'])
    .rename({'x':'x1','z':'target'},axis=1)
    .replace('a',0)
    .replace('b',1)
)

df.head()
```

```python
f, ax = plt.subplots(2,1,figsize=(9,9),sharex=True)

target_zero = df[df.target == 0].x1
target_one = df[df.target == 1].x1

ax[0].scatter(x=target_zero,y=[0 for _ in range(len(target_zero))],color='tab:blue',label='0')
ax[0].scatter(x=target_one,y=[0 for _ in range(len(target_one))],color='tab:orange',label='1')
ax[0].axhline(0,ls='--',xmin=0.01,xmax=0.99,color='black',alpha=0.2,zorder=0)
ax[0].set_yticks([])
ax[0].legend()

ax[1].boxplot(target_zero,positions=[0],vert=False)
ax[1].boxplot(target_one,positions=[1],vert=False)
ax[1].set_xlabel('x1')
ax[1].set_yticks([])

plt.show()
```

## Intuição (SVClassifier)

```python
MAX_ITER = 100

limites = []
erros = []

for iteration in range(MAX_ITER):
    ponto_zero = np.random.choice(target_zero,1)
    ponto_one = np.random.choice(target_one,1)
    
    limite = np.mean([ponto_zero,ponto_one])
    
                ## classificado zero ##     ## mas é 1 ##
    erro_zero = df.query("x1 < @limite").query("target == 1").count().target
    
                ## classificado one ##      ## mas é 0 ##
    erro_one = df.query("x1 > @limite").query("target == 0").count().target
    
    limites.append(limite)
    erros.append((erro_zero,erro_one))
else:
    erros = np.array(erros)
```

```python
plt.plot(erros[:,0]/len(target_zero))
plt.plot(erros[:,1]/len(target_one));
```
