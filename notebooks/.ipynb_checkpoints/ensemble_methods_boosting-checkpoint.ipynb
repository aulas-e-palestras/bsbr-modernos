# Gradient Boosting Machines (GBMs)

Gradient boosting machines é uma família de algoritmos de machine learning que, como as RandomForests (parentes distantes), utiliza um grupo de preditores fracos parar gerar um grande modelo mais forte. Usualmente esses modelos mais fracos são simples DecisionTrees e, normalmente, as GBM's geram resultados tão satisfatórios que são as queridinhas das competições do Kaggle e hackathons pelo país.

Mas o que são GBMs e como funcionam? Vamos responder essa perguntas nesse tutorial e, se tudo der certo, ao final você vai entender a lógica por trás de outros algoritmos que também usam estratégias parecidas.

## Importações

## Dados

```python
df = pd.read_csv('../data/regression_data_2.csv',usecols=['x','y'])
df.head()
```

```python
x,y = df.x,df.y

f, ax = plt.subplots(2,1,figsize=(11,9),sharex=True)

ax[0].scatter(x,y)
ax[0].set_ylabel('y')

ax[1].boxplot(x,vert=False)
ax[1].set_xlabel('x')
ax[1].set_yticks([])

plt.show()
```

## Intuição

Vamos super que temos um modelo que sempre chuta a média! Sua definição será:

$F_{0}(x) = \bar{y} $

```python
f, ax = plt.subplots(1,1,figsize=(11,6),sharex=True)
df.plot.scatter(x='x',y='y',ax=ax,s=50)
ax.axhline(df.y.mean(),c='tab:orange',label='média')
plt.legend();
```

Os resultados acima são claramente questionáveis e podemos ver avaliar esse resultado utilizando um gráfico de resíduos.

```python
f, ax = plt.subplots(1,1,figsize=(11,6),sharex=True)

media = df.y.mean()

residuos = [y-media for y in df.y]

ax.scatter(df.x,residuos)
ax.axhline(0,c='black',ls='--',alpha=0.2);
ax.set_title('Resíduos de f0');
```

Mas e se tivermos um outro modelo que prevê, na verdade, não o valor de y, mas sim os resíduos de $F_{0}$. Vamos ver como isso fica em um dataframe.

```python
df['res_0'] = residuos
df.head()
```

Então agora temos:

modelo    :   $h_{1}(x)$ <br>
features  :   $x$ <br>
target    :   $res_0 = y - F_{0}(x) $

```python
dt = DecisionTreeRegressor(max_depth=1,random_state=42).fit(df.x.values.reshape(-1,1),df.res_0)
```

```python
residuos_h1 = df.res_0 - dt.predict(df.x.values.reshape(-1,1))
df['res_1'] = residuos_h1
```

```python
df.head()
```

```python
f, ax = plt.subplots(1,1,figsize=(11,6),sharex=True)

plot_model_x = np.linspace(df.x.min(),df.x.max(),df.shape[0])

ax.scatter(df.x,df.res_0,label='residuos de f0',alpha=0.2)
ax.scatter(df.x,df.res_1,label='residuos de h1')
# ax.plot(plot_model_x,df.res_1,ls='--',c='tab:red',label='modelo h1')
ax.axhline(0,c='black',ls='--',alpha=0.2);
plt.legend();
```

Se juntarmos os valores preditos pelo modelo $F_0(x)$ (treinado para encontrar os valores $y$ ) aos resultados do modelo complementar $h_1(x)$ (treinado para encontrar os resíduos) teremos um modelo que entrega resultados melhores do que $F_0(x)$.

$F_{1}(x) = F_{0}(x) + h_{1}(x)$

```python
f, ax = plt.subplots(1,1,figsize=(11,6),sharex=True)


ax.scatter(df.x,df.y)
ax.axhline(df.y.mean(),c='tab:orange',label='f0')

f1_model = df.y.mean() + dt.predict(df.x.values.reshape(-1,1))

ax.plot(plot_model_x,f1_model,ls='--',c='tab:red',label='f1')

plt.legend();
```

## Dados

```python
year = [5,7,12,23,25,28,29,34,35,40]
salary = [82,80,103,118,172,127,204,189,99,166]

df = pd.DataFrame(data=zip(year,salary),columns=['year','salary'])
df.head()
```

Pontos importantes:

* Criaremos um modelo basaeado em árvore para tentar predizer salário.
* Nosso erro total (loss) será uma medida baseada na média dos quadrados dos resíduos (MSE).
* Nosso objetivo é encontrar uma função f(x) que minimize a loss definida.

Vamos tentar encontrar essa função...

## Primeira tentativa

```python
## tentando pela média

df = (
    df
    .assign(f0 = lambda x : x.salary.mean())             ## a média
    .assign(y_menos_f0 = lambda x: (x.salary - x.f0)**2) ## calculando o resíduo
)

df
```

```python
## construindo a árvore de regressão

random_split_point = np.random.randint(0,df.shape[0]-1)
random_split_point
```

```python
a = []

topo_do_split = df.loc[0:random_split_point,'y_menos_f0'].mean()
topo_do_split
```

```python
abaixo_do_split = df.loc[random_split_point:,'y_menos_f0'].mean()
abaixo_do_split
```

```python
novas_preds = [topo_do_split for _ in range(random_split_point)] + [abaixo_do_split for _ in range(df.shape[0] - random_split_point)]
```

```python
df['f1'] = novas_preds
```

```python
df['y_menos_f1'] = (df.salary - df.f1)**2
```

```python
df
```

```python
df.y_menos_f0.mean(), df.y_menos_f1.mean()
```

```python
x = df.drop(['target'],axis=1)
y = df.target

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```
